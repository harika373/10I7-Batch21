```python
'''Q. No. Question Expected Time
to complete
5
 Create a Line Chart Using Matplotlib.
 Plot CO2 Emissions Over Time
 Dataset Name: Global CO2 Emissions Dataset
(co2_emissions.csv)
 Objective: Track changes in CO2 emissions across years.
 Requirements:
o Load dataset with Year and CO2 Emissions.
o Plot emissions data using a line chart.
o Add labels and proper scaling.
 Deliverables:
o Python code.
o Line chart of CO2 emissions.
o Analysis of growth/reduction trends.
Expected Time
to Complete: 2
hours
'''
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset manually
data = {
    "Year": [2000, 2002, 2004, 2006, 2008, 2010,
             2012, 2014, 2016, 2018, 2020, 2022],
    "CO2_Emissions": [2500, 2600, 2800, 3000, 3300, 3500,
                      3700, 4000, 4200, 4100, 3900, 3950]
}

df = pd.DataFrame(data)

# Save to CSV (optional, for reuse later)
df.to_csv("co2_emissions.csv", index=False)

print(df)

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(df["Year"], df["CO2_Emissions"], marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel("Year")
plt.ylabel("CO2 Emissions (Million Metric Tons)")
plt.title("Global CO2 Emissions Over Time")

# Add grid for clarity
plt.grid(True)

plt.show()
```

```python
'''Q. No. Question Expected Time
to complete
6
 Create a Line Chart Using Matplotlib.
 Plot Student Performance Over Exams
 Dataset Name: Student Performance Dataset
(student_scores.csv)
 Objective: Visualize individual students score improvements.
 Requirements:
o Load dataset with Exam Number and Score.
o Plot line chart showing progression of scores.
o Add title and axis labels.
 Deliverables:
o Python code.
o Line chart of student performance.
o Insights about learning progress.
Expected Time
to Complete: 2
hours
'''
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset manually
data = {
    "Exam_Number": [1, 2, 3, 4, 5, 6],
    "Score": [55, 60, 68, 75, 82, 90]
}

df = pd.DataFrame(data)

# Save to CSV (optional, for reuse later)
df.to_csv("student_scores.csv", index=False)

print(df)

# Plot line chart
plt.figure(figsize=(8,5))
plt.plot(df["Exam_Number"], df["Score"], marker='o', linestyle='-', color='green')

# Add labels and title
plt.xlabel("Exam Number")
plt.ylabel("Score")
plt.title("Student Performance Over Exams")

# Show grid
plt.grid(True)

plt.show()
```
