```python
'''Question Expected Time
to complete
5
• Create a Bar Chart Using Matplotlib.
• Average Monthly Rainfall
• Dataset Name: Weather Dataset (monthly_rainfall.csv)
• Objective: Visualize rainfall variations across months.
• Requirements:
o Load dataset with Month and Rainfall (mm).
o Plot a bar chart with months on x-axis and rainfall
on y-axis.
o Use different bar colors for seasons.
• Deliverables:
o Python code.
o Rainfall bar chart.
o Insights about wettest and driest month.
Expected Time
to Complete: 2
hours'''
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (monthly_rainfall.csv should have columns: Month, Rainfall (mm))
try:
  df = pd.read_csv("monthly_rainfall.csv")
except FileNotFoundError:
  # Create a dummy dataset for demonstration if the file is not found
  data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
          'Rainfall (mm)': [50, 60, 80, 100, 120, 150, 180, 160, 130, 100, 70, 40]}
  df = pd.DataFrame(data)


# Define colors for seasons
season_colors = {
    'Winter': '#1f77b4',  # Blue
    'Spring': '#ff7f0e',  # Orange
    'Summer': '#2ca02c',  # Green
    'Autumn': '#d62728'   # Red
}

# Map months to seasons
month_to_season = {
    'Dec': 'Winter', 'Jan': 'Winter', 'Feb': 'Winter',
    'Mar': 'Spring', 'Apr': 'Spring', 'May': 'Spring',
    'Jun': 'Summer', 'Jul': 'Summer', 'Aug': 'Summer',
    'Sep': 'Autumn', 'Oct': 'Autumn', 'Nov': 'Autumn'
}

# Assign colors to each month based on season
colors = [season_colors[month_to_season[month]] for month in df['Month']]

# Plot Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.bar(df["Month"], df["Rainfall (mm)"], color=colors)

# Add labels and title
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.title("Average Monthly Rainfall")

# Add value labels above bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, round(yval, 2),
             ha='center', va='bottom', fontsize=10)

plt.show()

# Insights
wettest_month = df.loc[df['Rainfall (mm)'].idxmax()]
driest_month = df.loc[df['Rainfall (mm)'].idxmin()]

print(f"The wettest month is {wettest_month['Month']} with {wettest_month['Rainfall (mm)']:.2f} mm of rainfall.")
print(f"The driest month is {driest_month['Month']} with {driest_month['Rainfall (mm)']:.2f} mm of rainfall.")
```

```python
'''Question Expected Time
to complete
5
• Create a Bar Chart Using Matplotlib.
• Average Monthly Rainfall
• Dataset Name: Weather Dataset (monthly_rainfall.csv)
• Objective: Visualize rainfall variations across months.
• Requirements:
o Load dataset with Month and Rainfall (mm).
o Plot a bar chart with months on x-axis and rainfall
on y-axis.
o Use different bar colors for seasons.
• Deliverables:
o Python code.
o Rainfall bar chart.
o Insights about wettest and driest month.
Expected Time
to Complete: 2
hours'''
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (monthly_rainfall.csv should have columns: Month, Rainfall)
try:
  df = pd.read_csv("monthly_rainfall.csv")
except FileNotFoundError:
  # Create a dummy dataset for demonstration if the file is not found
  data = {'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
          'Rainfall': [50, 60, 80, 100, 120, 150, 180, 160, 130, 100, 70, 40]}
  df = pd.DataFrame(data)


# Define seasons and assign colors
season_colors = {
    "Winter": "#1f77b4",   # Blue
    "Summer": "#ff7f0e",   # Orange
    "Monsoon": "#2ca02c",  # Green
    "Autumn": "#d62728"    # Red
}

# Function to classify month into season (example for India)
def get_season(month):
    if month in ["December", "January", "February"]:
        return "Winter"
    elif month in ["March", "April", "May"]:
        return "Summer"
    elif month in ["June", "July", "August", "September"]:
        return "Monsoon"
    else:
        return "Autumn"

# Add season column
df["Season"] = df["Month"].apply(get_season)

# Plot bar chart with season colors
plt.figure(figsize=(10,6))
bars = plt.bar(df["Month"], df["Rainfall"],
               color=[season_colors[season] for season in df["Season"]])

# Add labels and title
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.title("Average Monthly Rainfall")

# Rotate x-axis labels for clarity
plt.xticks(rotation=45)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 3, int(yval),
             ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
