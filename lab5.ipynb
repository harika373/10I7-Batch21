```python
'''Merge Customers and Transactions Datasets and Compute
Spending Patterns
• Objective:
• Analyze customer purchase patterns by merging customer
and transaction datasets.
• Requirements:
• Create datasets (Customers and Transactions).
• Merge using customer_id.
• Calculate total spend per customer.
• Group by city to analyze spending.
• Identify top spender.
• Deliverables:
• Python code for dataset creation, merging, and spending
analysis.
• Output showing merged dataset, customer spend, and city-
wise spend.
• Insights into customer behavior and city-wise spending
patterns.
Expected Time
to Complete: 2
hours'''
import pandas as pd

# ------------------ Step 1: Create Datasets ------------------
# Customers dataset
customers_data = {
    "customer_id": [1, 2, 3, 4, 5],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "city": ["Delhi", "Mumbai", "Delhi", "Bangalore", "Mumbai"]
}
customers_df = pd.DataFrame(customers_data)

# Transactions dataset
transactions_data = {
    "transaction_id": [101, 102, 103, 104, 105, 106, 107],
    "customer_id": [1, 2, 1, 3, 4, 2, 5],
    "amount": [250, 400, 300, 150, 500, 200, 350]
}
transactions_df = pd.DataFrame(transactions_data)

# ------------------ Step 2: Merge Datasets ------------------
merged_df = pd.merge(transactions_df, customers_df, on="customer_id", how="inner")

print("Merged Dataset:")
print(merged_df)

# ------------------ Step 3: Compute Spending ------------------
# Total spend per customer
customer_spend = merged_df.groupby(["customer_id", "name"])["amount"].sum().reset_index()

print("\nTotal Spend Per Customer:")
print(customer_spend)

# ------------------ Step 4: Group by City ------------------
city_spend = merged_df.groupby("city")["amount"].sum().reset_index()

print("\nCity-Wise Spending:")
print(city_spend)

# ------------------ Step 5: Identify Top Spender ------------------
top_spender = customer_spend.loc[customer_spend["amount"].idxmax()]

print("\nTop Spender:")
print(top_spender)
```

```python
'''Merge Patients and Appointments Datasets and Compute
Healthcare Statistics
• Objective:
• Merge patient and appointment datasets to analyze hospital
visits and doctor workloads.
• Requirements:
• Create datasets (Patients and Appointments).
• Merge using patient_id.
• Count appointments per patient.
• Group by doctor to compute workload.
• Find patient with maximum visits.
• Deliverables:
• Python code to merge and analyze healthcare datasets.
• Output tables showing patient visits and doctor workload.
• Insights into patient frequency and doctor distribution.
Expected Time
to Complete: 2
hours'''
import pandas as pd

# ------------------ Step 1: Create Datasets ------------------
# Patients dataset
patients_data = {
    "patient_id": [1, 2, 3, 4, 5],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "age": [25, 40, 35, 60, 50],
    "city": ["Delhi", "Mumbai", "Delhi", "Bangalore", "Mumbai"]
}
patients_df = pd.DataFrame(patients_data)

# Appointments dataset
appointments_data = {
    "appointment_id": [101, 102, 103, 104, 105, 106, 107, 108],
    "patient_id": [1, 2, 1, 3, 4, 2, 5, 1],
    "doctor": ["Dr. A", "Dr. B", "Dr. A", "Dr. C", "Dr. B", "Dr. B", "Dr. C", "Dr. A"],
    "date": ["2025-09-01", "2025-09-02", "2025-09-05", "2025-09-06",
             "2025-09-08", "2025-09-10", "2025-09-12", "2025-09-15"]
}
appointments_df = pd.DataFrame(appointments_data)

# ------------------ Step 2: Merge Datasets ------------------
merged_df = pd.merge(appointments_df, patients_df, on="patient_id", how="inner")

print("Merged Dataset:")
print(merged_df)

# ------------------ Step 3: Count Appointments per Patient ------------------
patient_visits = merged_df.groupby(["patient_id", "name"])["appointment_id"].count().reset_index()
patient_visits.rename(columns={"appointment_id": "total_visits"}, inplace=True)

print("\nAppointments Per Patient:")
print(patient_visits)

# ------------------ Step 4: Group by Doctor (Workload) ------------------
doctor_workload = merged_df.groupby("doctor")["appointment_id"].count().reset_index()
doctor_workload.rename(columns={"appointment_id": "total_appointments"}, inplace=True)

print("\nDoctor Workload:")
print(doctor_workload)

# ------------------ Step 5: Find Patient with Maximum Visits ------------------
top_patient = patient_visits.loc[patient_visits["total_visits"].idxmax()]

print("\nPatient with Maximum Visits:")
print(top_patient)
```
